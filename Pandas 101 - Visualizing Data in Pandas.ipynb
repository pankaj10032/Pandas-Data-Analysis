# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\panka\Desktop\Exploratory_data_analysis\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available) # to see different available styles that we can use
plt.style.use('classic')
```

```python
df.plot() # plot of all the columns
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='bar', xlabel='Daily rating', ylabel='scores')
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked = True)  # bar plot stacked on each other  
# barh-> for horizontal bar plot
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Yellow') # s-> to change the size of dots here, c-> color of each point
```

```python
df.plot.hist(bins=20)
```

```python
df.plot.hist(bins = 10) # histogram is for showing the disrtibution of numbers
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10)) # bar chat can only be made for a particular columns so we have to specify what column we are working with
```
